```python
flipkart_url = "https://www.flipkart.com/search?q=iphone" #preparing the URL to search the product on flipkart
uClient = uReq(flipkart_url) # requesting the webpage from the internet
flipkartPage = uClient.read() # reading the webpage it means that we have downloaded the HTML content
uClient.close() # closing the connection to the web server after we have downloaded the content
```

```python
# once the entire HTML has been downloaded we will write the code to understand the HTML completely.
# we use beautiful soup library in order to understand the content. "html.parser" tells the beautiful soup that the content is
# an HTML content.
flipkart_html = bs(flipkartPage, "html.parser") # parsing the webpage as HTML
flipkart_html
```

```python
bigboxes = flipkart_html.findAll("div", {"class": "bhgxx2 col-12-12"}) 
# seacrhing for appropriate tag to redirect to the product link beautiful soup 
```

```python
del bigboxes[0:2] # the first 3 members of the list do not contain relevant information, hence deleting them.
box = bigboxes[0] #  taking the first iteration (for demo)
```

```python
productLink = "https://www.flipkart.com" + box.div.div.div.a['href'] # extracting the actual product link
# we need to understand here that we have to concatenate the flipkart link before the href link. this is because the href link
# can't take us to the next page individually.
```
