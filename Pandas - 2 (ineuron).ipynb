```python
import numpy as np
import pandas as pd
from numpy.random import randn as rn
np.random.seed(101)  # random function helps to generate random numbers on standard deviation.
matrix_data = rn(5,4)
row_labels = ['A','B','C','D','E']
column_headings = ['W','X','Y','Z']

df = pd.DataFrame(data=matrix_data, index=row_labels, columns=column_headings)
print("\nThe data frame looks like\n",'-'*45, sep='')
print(df)
```

```python
print(df.X)
```

```python
print("\nA column is created by assigning it in relation to an existing column\n",'-'*75, sep='')
df['New'] = df['X']+df['Z']
df['New (Sum of X and Z)'] = df['X']+df['Z']
print(df)
```

```python
# when we have to remove the row we keep axis = 0, whereas when we have to remove the column we have to keep axis = 1
# there is command called 'inplace' which is false by default. that means that if we remove the column, it doesn't remove
# the column from the original dataframe. However if you write inplace = True then the command will remove the entire column
# permanently.
print("\nA column is dropped by using df.drop() method\n",'-'*55, sep='')
df = df.drop('New', axis=1) # Notice the axis=1 option, axis = 0 is default, so one has to change it to 1
print(df)
```

```python
# del command deletes data from column whereas drop helps to delete from both row and column. 
# 
```
